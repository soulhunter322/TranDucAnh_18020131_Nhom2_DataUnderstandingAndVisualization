```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
mat = pd.read_json('vnexpress3_20200803_015519.txt',lines=True)
mat.dropna()
mat
```

```python
category_name = []
category_count  = []
for x in mat.category:
    value = x
    if value in category_name: category_count[category_name.index(value)]+=1 
    else: 
        category_name.append(value)
        category_count.append(1)

category = list(zip(category_name,category_count))
dp = pd.DataFrame(data=category, columns=['Loại', 'số lượng'])
dp
```
