```python
import pandas as pd
import matplotlib.pyplot as plt     
import matplotlib.ticker as ticker
import numpy as np
import seaborn as sns
```

```python
plt.rcParams['figure.figsize'] = (8,8)
prd = pd.read_csv('TGDD.csv')
prd
```

```python
prd.Rate = (prd.Rate - 5)/100 
prd
```

```python
type_name = []
type_count  = []
for x in prd.Type:
    value = x
    if value in type_name: type_count[type_name.index(value)]+=1 
    else: 
        type_name.append(value)
        type_count.append(1)

type = list(zip(type_name,type_count))
df = pd.DataFrame(data=type, columns=['type', 'count'])
df
```

```python
print('số lượng sản phẩm cao nhất: ',df['count'].max())
print('số lượng sản phẩm thấp nhất: ',df['count'].min())
print('số lượng sản phẩm trung bình: ',df['count'].mean())
```

```python
plt.pie(df['count'], labels = df['type'])
plt.show()
```

```python
g = sns.JointGrid(x="Current Price", y="Rate", data=prd)
fig=plt.gcf()
fig.set_size_inches(10,10)
g = g.plot(sns.regplot, sns.distplot)
```

```python
plt.xlabel('số lượng')
plt.ylabel('loại sản phẩm')
plt.suptitle('Tương quan số lượng bán hàng giữa các loại sản phẩm')
plt.barh('type','count', color='red',data = df)
plt.show()
```

```python
bodies_by_type = prd.groupby('Type').sum()
bodies_by_type
```

```python
print('thành phần doanh thu cho từng loại sản phẩm:')
plt.pie(bodies_by_type['Current Price'], labels = df['type'])
plt.show()
```
